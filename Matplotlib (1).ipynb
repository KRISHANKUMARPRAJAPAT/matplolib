## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


# Ans:1


Matplotlib is a popular data visualization library in Python that allows users to create a wide variety of static, animated, and interactive visualizations in Python.

Matplotlib is used for creating charts, graphs, and other types of visualizations from data. It provides a range of functions for creating different types of plots, such as line plots, scatter plots, bar charts, histograms, and many more.

**Some of the most commonly used plots that can be created using the Pyplot module of Matplotlib include:**

**1.Line Plot:** A plot where data points are represented by points or symbols that are connected by lines. It is useful for visualizing trends over time.

**2.Scatter Plot:** A plot where data points are represented as points on a graph, where the x-axis and y-axis represent different variables. It is useful for identifying patterns in data.

**3.Bar Chart:** A plot where data is represented using vertical or horizontal bars. It is useful for comparing data between different categories.

**4.Histogram:** A plot where data is represented as bars that indicate the frequency of values within a certain range. It is useful for visualizing the distribution of data.

**5.Pie Chart:** A plot where data is represented as a circular chart divided into segments, where each segment represents a category. It is useful for visualizing the proportions of different categories in a dataset.


## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

**Note: Also add title, xlabel, and ylabel to the plot.**

# Ans: 2

A scatter plot is a type of plot or graph that displays the relationship between two variables in a dataset. It is used to determine whether there is any correlation or relationship between the variables, and to identify any patterns or trends in the data.

In a scatter plot, each data point is represented by a dot or symbol on the graph, with the x-axis and y-axis representing the two variables being compared. The position of each dot on the graph indicates the value of the two variables for that particular data point.Scatter plots are commonly used in fields such as statistics, data analysis, and data visualization to explore relationships between variables and to identify any patterns or trends in the data.

```python
import matplotlib.pyplot as plt 
import numpy as np 

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.title("Scatter Plot of X and Y ")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# Ans: 3 


The subplot() function in Matplotlib is used to create multiple subplots within a single figure. This is useful when you want to display multiple plots side-by-side or in a grid, and can help to provide a more organized and efficient way to display information.
Using the subplot() function can be especially useful when creating complex visualizations or when comparing multiple sets of data. It allows you to display multiple plots in a single figure, which can help to make your visualizations more clear and informative. Additionally, it can save space and make your code more concise by reducing the number of figures you need to create.

```python
# crate data for line 1;
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# create data for line 2 ;
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# create data for line 3 ;
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# create data for line 4 ;
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create a 2x2 grid of subplots
fig, axs = plt.subplots(2,2)

# create the first plot in the top-left subplot
axs[0,0].plot(x1, y1)
axs[0,0].set_title('Line 1')

# create the second plot in the top-right subplot
axs[0,1].plot(x2, y2)
axs[0,1].set_title('Line 2')

# create the third plot in the bottom-left subplot
axs[1,0].plot(x3, y3)
axs[1,0].set_title('Line 3')

# create the forth plot in the bottom-right subplot
axs[1,1].plot(x4,y4)
axs[1,1].set_title('Line 4')

fig.suptitle('Four Line Plots')
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

# Ans: 4 


A bar plot, also known as a bar chart, is a type of plot that displays categorical data with rectangular bars. Each bar represents a category, and the height or length of the bar corresponds to the value or count of that category. Bar plots are used to compare the relative sizes or frequencies of different categories or to show changes in those values over time.

Bar plots are used in a variety of fields such as economics, finance, social sciences, and data analysis, to represent data such as stock prices, GDP, or population demographics. They can be used to identify trends, visualize distributions, or compare different categories in a dataset.

Bar plots are also commonly used in data visualization and communication to present information in a clear and concise manner. They can help to convey complex information in a simple and understandable way, making it easier for viewers to interpret and draw insights from the data.

```python
company = np.array(['Apple' , "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# create a vartical bar plot 
plt.bar(company , profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ( in Millions)')
plt.show()

# creatimg a horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

# Ans: 5

A box plot is a graphical representation of a set of data that displays the distribution of values through their quartiles. It is also known as a box and whisker plot, and it is used to summarize and visualize the central tendency, variability, and skewness of a dataset.

A typical box plot consists of a box that represents the interquartile range (IQR), which is the range between the first and third quartiles, and a vertical line inside the box that represents the median (second quartile). The whiskers extend from the box to the minimum and maximum values that fall within a certain range (usually 1.5 times the IQR), and any data points that lie outside of this range are considered outliers and plotted as individual points.

Box plots are useful because they provide a quick and easy way to compare the distribution of different datasets or to examine the distribution of a single dataset. They can also be used to detect outliers and to identify any skewness in the data. Box plots are commonly used in statistical analysis, data visualization, and in scientific research.
